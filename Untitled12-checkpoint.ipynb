```python
import pandas as pd
import pandas as pd
file_path = 'Earthquakes_database - Earthquakes_database.csv'
df = pd.read_csv(file_path)
df
```

```python
#წავშალოთ სვეტი Root Mean Square
df = df.drop(columns = "Root Mean Square")
df
```

```python
#შევცვალოთ თითოეული ცარიელი მნიშვნელობა 'Unknown'-ით
df.fillna('Unknown', inplace=True)
df = df.astype('object')
print("\nDataFrame with Empty Values Replaced:")
print(df)
```

```python
#წავშალოთ Whitespace-ები Source სვეტიდან
df['Source'] = df['Source'].str.strip()
print("\nDataFrame with Whitespaces Removed:")
print(df)
```

```python
#გავიგოთ მინიმალური მაგნიტუდის მნიშვნელობა ცხრილში
min_magnitude = df['Magnitude'].min()
print(min_magnitude)
```

```python
#გავიგოთ მაქსიმალური მაგნიტუდის მნიშვნელობა
max_magnitude = df['Magnitude'].max()
print(max_magnitude)
```

```python
#დავყოთ მაგნიტუდები კატეგორიებად 'Low', 'Average', 'High' და შევქმნათ ახალი სვეტი, სახელად 'Magnitude Category'
bins = [0,5.5,7,9.1]
labels = ['Low', 'Average', 'High']
df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, include_lowest=True)
print(df['Magnitude Category'])
```

```python

#გამოვიყენოთ stack() სვეტების სტრიქონებად გადასაქცევად. 
stacked_df = df.set_index('ID').stack()
print("Stacked DataFrame:")
print(stacked_df)
```

```python
#გამოვიყენოთ unstack() სტრიქონების სვეტებად გადასაქცევად.
unstacked_df = stacked_df.unstack()
print("\nUnstacked DataFrame:")
print(unstacked_df)
```

```python
#კვადრატული ფესვის წესის გამოყენებით ბინის რაოდენობის დადგენა
import math
num_bins = int(math.sqrt(len(df)))

print("Number of bins (square root rule):", num_bins)
```

```python
#ჰისტოგრამის აგება გრძედისთვის

import matplotlib.pyplot as plt

plt.hist(df['Longitude'], bins=153, color = 'red', edgecolor='black')
plt.title('Histogram of Longitude')
plt.xlabel('Longitude')
plt.ylabel('Frequency')
plt.show()
```

```python
#გაფანტულობის დიაგრამის აგება განედისთვის
plt.scatter(df.index, df['Latitude'], alpha=0.5, color = '#f752ac')
plt.title('Scatter Plot for {}'.format('Latitude'))
plt.xlabel('Index')
plt.ylabel('Latitude')
plt.show()
```

```python
#ხაზოვანი დიაგრამის აგება სიღრმისთვის
plt.plot(df.index, df['Depth'], label='Depth Line Chart', color='#00fff7')
plt.title('Line Chart for {}'.format('Depth'))
plt.xlabel('Index')
plt.ylabel('Depth')
plt.legend()
plt.show()
```

```python
#სვეტოვანი დიაგრამის აგება მაგნიტუდისთვის
plt.plot(df.index, df['Magnitude'], label='Depth Bar Chart', color='#750b9c')
plt.title('Bar Chart for {}'.format('Magnitude'))
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
```
